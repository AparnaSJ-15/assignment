```python
import pandas as pd
import sqlite3
conn=sqlite3.connect('test_database')
c=conn.cursor()
c.execute('CREATE TABLE IF NOT EXISTS product1(API WELL NUMBER number, OIL number)')
conn.commit()
c.execute('CREATE TABLE IF NOT EXISTS product2(API WELL NUMBER number, GAS number )')
conn.commit()
c.execute('CREATE TABLE IF NOT EXISTS product3(API WELL NUMBER number, BRINE number )')
conn.commit()
```

```python
df1=pd.read_excel("20210309_2020_1 - 4 (1).xls",index_col=0,header=0,usecols=[0,2,8,9,10])
```

```python
print(df1)
```

```python
o=pd.DataFrame(df1.groupby(['API WELL  NUMBER'])['OIL'].sum().reset_index())
o.to_sql('product1', conn, if_exists='replace', index = False)
```

```python
c.execute('''  
SELECT * FROM product1
          ''')

for row in c.fetchall():
    print (row)
```

```python
g=pd.DataFrame(df1.groupby(['API WELL  NUMBER'])['GAS'].sum().reset_index())
g.to_sql('product2', conn, if_exists='replace', index = False)
```

```python
c.execute('''  
SELECT * FROM product2
          ''')

for row in c.fetchall():
    print (row)
```

```python
b=pd.DataFrame(df1.groupby(['API WELL  NUMBER'])['BRINE'].sum().reset_index())
b.to_sql('product3', conn, if_exists='replace', index = False)
```

```python
c.execute('''  
SELECT * FROM product3
          ''')

for row in c.fetchall():
    print (row)
```
